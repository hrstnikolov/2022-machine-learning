# Tree and Ensemble Methods Demos

```python
relevant_columns = [
#     'instant',
#     'dteday',
    'season',
#     'yr',
#     'mnth',
#     'hr',
    'holiday',
#     'weekday',
    'workingday',
    'weathersit',
    'temp',
    'atemp',
    'hum',
    'windspeed',
    'casual',
    'registered',
    'cnt'
]

new_names = [
    'season',
    'holiday',
    'workingday',
    'weathersit',
    'temp',
    'atemp',
    'hum',
    'windspeed',
    'casual',
    'registered',
    'cnt'
]

t_max = 39
t_min = -8
```

```python
bikes = pd.read_csv("../datasets/bikes/day.csv")
timestamp = pd.to_datetime(bikes.dteday)
bikes = bikes.set_index(timestamp)
bikes = bikes[relevant_columns]
bikes["real_temp"] = bikes.temp * (t_max-t_min) + t_min

bikes.head()
```

```python
plt.plot(bikes.real_temp)

plt.xlabel("Time")
plt.ylabel("Temperature C")
plt.title("Average temperature 2011-2012")
plt.show()
```

```python
plt.plot(bikes.real_temp, bikes.casual, 'o')

plt.xlabel("Temperature [C]")
plt.ylabel("Count of rented bikes")
plt.title("Rented bikes per day")
plt.show()
```

### Linear regression

```python
X = bikes.real_temp.values.reshape(-1, 1)
y = bikes.casual.values
```

```python
reg = LinearRegression()
reg.fit(X, y)


print(f"Model line: y = {reg.coef_[0]:.1f}x + {reg.intercept_:.1f}")

bikes_when_5deg = int(reg.predict(np.array(5).reshape(-1,1)))
print(f"How many bikes will be hired when the temperature is 5 degrees? Answer: {bikes_when_5deg}")


x_plot = [-5, 30]
y_plot = reg.coef_ * x_plot + reg.intercept_

plt.plot(bikes.real_temp, bikes.casual, 'o')
plt.plot(x_plot, y_plot, 'r')

plt.xlabel("Temperature [C]")
plt.ylabel("Count of rented bikes")
plt.title("Fitted Linear Regression")
plt.show()
```

### Decision tree

```python
regressor = DecisionTreeRegressor(max_depth=2)
regressor.fit(X, y)

print(f"feature importances: {regressor.feature_importances_}")
print(f"{regressor.predict(np.array(5).reshape(-1,1))}")

plot_tree(regressor)
plt.show()
```

```python
# decision tree
xx = np.array([np.linspace(-5, 30, 100)]).T
yy = regressor.predict(xx)
plt.plot(xx, yy, 'r', linewidth=5, label="Prediction by decision tree", alpha=0.5)

# linear regression
plt.plot(x_plot, y_plot, 'g', linewidth=5, label="Prediction by linear regression", alpha=0.5)

# data points
plt.plot(bikes.real_temp, bikes.casual, '.', alpha=0.3)


plt.xlabel("Temperature [C]")
plt.ylabel("Count of rented bikes")
plt.title("Fitted Decision Tree Regressor")
plt.legend()
plt.show()
```

```python
print("R2 scores")

print("Dcission tree: \t\t", r2_score(y, regressor.predict(X)).round(2))
print("Linear Regression: \t", r2_score(y, reg.predict(X)).round(2))
```

## Impurity measures

  

```python
def entropy(p):
    return -p * np.log2(p) - (1 - p) * np.log2((1 - p))
def gini_index(p):
    return p * (1 - p) + (1 - p) * (1 - (1 - p))
def misclassification_error(p):
    return 1 - np.max([p, 1 - p])
```

```python
prob_of_class_one = np.linspace(0, 1, 100)
```

```python
plt.plot(prob_of_class_one, entropy(prob_of_class_one), label = "entropy")
plt.plot(prob_of_class_one, 0.5 * entropy(prob_of_class_one), label="x0.5 scaled entropy")
plt.plot(prob_of_class_one, gini_index(prob_of_class_one), label="gini index")
# plt.plot(prob_of_class_one, misclassification_error(prob_of_class_one))

plt.gca().set_aspect("equal")
plt.xlabel("P(X=1) - Probability of class 1 (the positive class)")
plt.ylabel("Impurity metric")
plt.legend()
plt.show()
```

## Visualizing decision tree boundaries

```python
iris = load_iris()
X = iris.data[:, :2]
y = iris.target
```

```python
color_dict = {0: "blue", 1: "lightgreen", 2: "red"}
colors = [color_dict[i] for i in y]

depth_1 = DecisionTreeClassifier(max_depth = 1).fit(X, y)
depth_2 = DecisionTreeClassifier(max_depth = 2).fit(X, y)
depth_3 = DecisionTreeClassifier(max_depth = 3).fit(X, y)
depth_4 = DecisionTreeClassifier(max_depth = 4).fit(X, y)
titles = ["Max depth = 1","Max depth = 2","Max depth = 3", "Max depth = 4"]

h = 0.02 # Step size

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

for i, classifier in enumerate((depth_1, depth_2, depth_3, depth_4)):
    plt.figure(figsize=(4,4))
    Z = classifier.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap = plt.cm.coolwarm, alpha = 0.8)
    plt.scatter(X[:, 0], X[:, 1], c = colors)
    
    plt.xlabel("Sepal length")
    plt.ylabel("Sepal width")
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.xticks(())
    plt.yticks(())
    plt.title(titles[i])
plt.show()
```

```python
plt.figure(figsize=(12,12))
plot_tree(depth_4)
plt.show()
```

## Housing

https://archive.ics.uci.edu/ml/machine-learning-databases/housing/

- CRIM - per capita crime rate by town
- ZN - proportion of residential land zoned for lots over 25,000 sq.ft.
- INDUS - proportion of non-retail business acres per town.
- CHAS - Charles River dummy variable (1 if tract bounds river; 0 otherwise)
- NOX - nitric oxides concentration (parts per 10 million)
- RM - average number of rooms per dwelling
- AGE - proportion of owner-occupied units built prior to 1940
- DIS - weighted distances to five Boston employment centres
- RAD - index of accessibility to radial highways
- TAX - full-value property-tax rate per USD 10,000
- PTRATIO - pupil-teacher ratio by town
- B - 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
- LSTAT - % lower status of the population
- MEDV - Median value of owner-occupied homes in USD 1000's

## Boosting

### Ada Boosting

```python
column_names = [
    "age",
    "workclass",
    "fnlwgt",
    "education",
    "education-num",
    "marital-status",
    "occupation",
    "relationship",
    "race",
    "sex",
    "capital-gain",
    "capital-loss",
    "hours-per-week",
    "native-country",
    "income_class",
]
```

```python
directory = Path("../datasets/adult/")
income_train = pd.read_csv(directory / "adult.data", header=None)
income_test = pd.read_csv(directory / "adult.test", header=None, skiprows=1)

income_data = pd.concat([income_train, income_test])

income_data.columns = column_names
income_data.income_class = income_data.income_class.str.strip()

income_data.shape
```

```python
income_attributes = income_data.drop(columns="income_class")
income_target = income_data.income_class

income_attributes = pd.get_dummies(income_attributes)

scaler = MinMaxScaler()
income_attributes = scaler.fit_transform(income_attributes)

X_train, X_test, y_train, y_test = \
train_test_split(income_attributes, income_target, train_size=0.7)

for split_data in X_train, X_test, y_train, y_test:
    print(split_data.shape)
```

```python
# Single tree
tree = DecisionTreeClassifier(max_depth = 1)
tree.fit(X_train, y_train) 
train_pred = accuracy_score(y_train, tree.predict(X_train))
test_pred = accuracy_score(y_test, tree.predict(X_test))
print(f"Decision tree train / test accuracies: {train_pred: .3f} / {test_pred: .3f}")
```

```python
# Boosted tree
ada = AdaBoostClassifier(base_estimator = tree, n_estimators = 30, learning_rate = 0.1)
ada.fit(X_train, y_train)
```

```python
train_pred = accuracy_score(y_train, ada.predict(X_train))
test_pred = accuracy_score(y_test, ada.predict(X_test))
print(f"AdaBoost tree train / test accuracies: {train_pred: .3f} / {test_pred: .3f}")
```
