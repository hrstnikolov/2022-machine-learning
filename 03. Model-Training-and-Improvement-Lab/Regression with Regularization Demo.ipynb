```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression, Lasso, Ridge, RidgeCV, ElasticNet
```

```python
RANDOM_STATE = 1
```

# Regression with Regularization Demo

## Generated data

### Single feature

```python
x, y = make_regression(
    n_samples=100,
    n_features=1,
    noise = 40,
    random_state=RANDOM_STATE,
)

plt.scatter(x, y)
plt.show()
```

```python
lr = LinearRegression()
lr_l1 = Lasso(alpha=1)
lr_l2 = Ridge(alpha=1)
lr_both = ElasticNet(alpha = 1.0, l1_ratio = 0.5)
models = [lr, lr_l1, lr_l2, lr_both]

for model in models:
    model.fit(x, y)
    
descriptions = [
    "No regularization",
    "L1",
    "L2",
    "L1 and L2",
]

names = [
    "LinearRegression",
    "Lasso",
    "Ridge",
    "ElasticNet",
]
```

```python
for model, name, description in zip(models, names, descriptions):
    R_sq = model.score(x, y)
    print(f"{name} R_sq = {R_sq:.2f}")
```

### Multiple features

```python
X, y = make_regression(
    n_samples=5000,
    n_features=10,
    noise = 20,
    random_state=RANDOM_STATE,
)

plt.scatter(X[:,0], y, marker='.')
# plt.scatter(X[:,1], y)
# plt.scatter(X[:,2], y)
plt.show()
```

```python
lr = LinearRegression()
lr_l1 = Lasso(alpha=1)
lr_l2 = Ridge(alpha=1)
lr_both = ElasticNet(alpha = 1.0, l1_ratio = 0.5)
models = [lr, lr_l1, lr_l2, lr_both]

for model in models:
    model.fit(X, y)
    
descriptions = [
    "No regularization",
    "L1",
    "L2",
    "L1 and L2",
]

names = [
    "LinearRegression",
    "Lasso",
    "Ridge",
    "ElasticNet",
]
```

```python
for model, name, description in zip(models, names, descriptions):
    R_sq = model.score(X, y)
    print(f"{name} R_sq = {R_sq:.2f}")
```
