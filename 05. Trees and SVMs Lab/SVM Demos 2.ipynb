```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.svm import SVC, LinearSVC
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, f1_score, classification_report, make_scorer
```

# SVM Demos 2

```python
X, y = make_blobs(
    n_samples=20,
    n_features=2,
    centers=[[4, 1], [1, 4]],
    cluster_std=1.0,
    random_state=42,
)

def plot_blobs(X, y, marker='o', alpha=1):
    plt.scatter(X[:, 0], X[:, 1], c=y, marker=marker, alpha= alpha)
    plt.xlabel('$X_0$')
    plt.ylabel('$X_1$')
    
plot_blobs(X, y)
```

```python
lsvc = LinearSVC()
lsvc.fit(X, y)
```

```python
print(lsvc.predict([[1, 2], [2, 5], [3, 6]]))  # expected all 1
print(lsvc.predict([[4, -1], [2, 0], [5, 2]]))  # expected all 0
```

```python
# # AttributeError
# lsvc.support_ 
```

```python
svc = SVC(kernel='linear')
svc.fit(X, y)
support_vector_indeces = svc.support_
support_vector_indeces
```

```python
plot_blobs(X, y, marker='.')
plot_blobs(X[support_vector_indeces,:], y[support_vector_indeces], marker='X')
```

```python
# points to plot the boundery areas
h = 0.1
x1, x2 = np.meshgrid(np.arange(0, 5, h), np.arange(-1, 6, h))
points = np.vstack([x1.ravel(), x2.ravel()]).T
predictions = svc.predict(points)
```

```python
plot_blobs(X, y, marker='o')
plot_blobs(points, predictions, marker='.', alpha=0.1)
plot_blobs(X[support_vector_indeces,:], y[support_vector_indeces], marker='X')
plt.xlim(0, 5)
plt.ylim(-1, 6)
plt.show()
```
